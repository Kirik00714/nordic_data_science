4. Различается ли отток по штатам? Можно ли добавить эту "фичу" в нашу прогнозную модель? Имеет ли это смысл?

```python
i = 0
for row in df.values:
    sum_minutes = row[6] + row[9] + row[12]
    df.loc[i,'Sum minutes'] = sum_minutes 
    i += 1
```

```python
df[['State','Sum minutes']].sort_values(by = 'Sum minutes',ascending=False).head(3)
```

```python
df[['State','Sum minutes']].sort_values(by = 'Sum minutes',ascending=False).head(5)
```

```python
df_2 = df[['State','International plan','Sum minutes']].sort_values(by = 'Sum minutes',ascending=False)
df_2=df_2[df_2['International plan'] == 'Yes']
pd.crosstab(df_2['State'], df_2['International plan'], normalize=True)*100
```

```python
pd.crosstab(df['Churn'], df['State'] , normalize=True)*100
```

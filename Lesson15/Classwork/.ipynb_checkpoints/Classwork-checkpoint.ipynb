```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
def get_html(url: str) -> str:
    responce = requests.get(url)
    if responce.status_code == 200:
        return responce.text
    else:
        return None
```

```python
url = 'http://hyaenidae.narod.ru/story1/038.html'
html = get_html(url)
```

```python
if not html is None:
    soup = BeautifulSoup(html)
else:
    print('Страница не загрузилась')
```

```python
def get_info_from_html(soup: BeautifulSoup) -> tuple:
    header = soup.find('div', class_ = 'rus')
    if not header is None:
        name_of_tale = header.find('h1').text
    else:
        print("Заголовок не найден")
    p_sentences = soup.find('td', class_ = 'textskazki').find_all('p')
    tale_paragraphs = [p.text for p in p_sentences]
    tale_text = ' '.join(tale_paragraphs)
    return name_of_tale, tale_paragraphs,tale_text
```

```python
def get_tale(url: str)-> tuple:
    html = get_html(url)
    if not html is None:
        soup = BeautifulSoup(html)
    else:
        print('Страница не загрузилась')
        return None
    return get_info_from_html(soup)
```

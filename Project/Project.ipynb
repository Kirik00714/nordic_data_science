```python
text = requests.get('https://autospot.ru/').text

soup = BeautifulSoup(text)

domain = 'https://autospot.ru'

list_name_car = []
for i in tqdm(soup.find_all(class_ = 'car-search__inner')):
    buf_soup = i.find_all(class_ = 'car-search__inner-col')
    for j in buf_soup:
        t = j.find_all('a')
        for k in t:
            z = k.get('href')
            list_name_car.append(domain+z)
```

```python
list_topic = []
for name_car in tqdm(list_name_car):
    text = requests.get(name_car).text

    soup = BeautifulSoup(text)

    domain = 'https://autospot.ru'

    for i in soup.find_all(class_ = 'car-model-item__description'):
        buf_soup = i.find('a').get('href')
        list_topic.append(domain+buf_soup)
```

```python
model_car = []
for item in tqdm(list_topic):
    text = requests.get(item).text
    soup_2 = BeautifulSoup(text)
    for i in soup_2.find_all(class_ = 'auto-card__top-line-block'):
        model_car.append(i.find('a').get('href'))
```

```python
toplivo = []
obem_dvs = []
hs = []
KPP = []
stupni = []
privod = []
smesh = []
city = []
road = []
do100 = []
year = []
for data in tqdm(dataset_from_file):
    regex_num = re.compile('([-+]*\d+\.\d+|[-+]*\d+)') 
    try:
        if data[0] == 'Бензин' or data[0] == 'Дизель' or data[0] == 'Электромотор' or data[0] == 'Газ':
            toplivo.append(data[0])
    except:
        print('Ошибка занесения в список топливо')
    try:
        if ' ' in data[1]:
            obem_dvs.append(data[1][1:])
    except:
        print('Ошибка занесения в список объем двс')
        
    try:
        if 'л.с.' in data[1] :
            hs.append(data[1][1:4])
        if 'л.с.' in data[2]:
            hs.append(data[2][1:4])
    except:
        print('Ошибка занесения в список лошадиные силы')

    try:
        if 'Автомат' in data[2] or 'Робот' in data[2] or 'Механика' in data[2] or 'Вариатор' in data[2]:
            KPP.append(data[2])
        if 'Автомат' in data[3] or 'Робот' in data[3] or 'Механика' in data[3] or 'Вариатор' in data[3]:
            KPP.append(data[3])
    except:
        print('Ошибка занесения в список КПП')

    try:
        if 'ступеней' in data[4]:
            stupni.append(data[4][1:2])
    except:
        print('Ошибка занесения в список ступеней в КПП')

    try:
        if 'Полный' in data[2] or 'Передний' in data[2] or 'Задний' in data[2]:
            privod.append(data[2])
        if 'Полный' in data[5] or 'Передний' in data[5] or 'Задний' in data[5]:
            privod.append(data[5])
    except:
        print('Ошибка занесения в список привод')
        
    try:
        if 'смешанный' in data[6]:
            smesh.append(regex_num.findall(data[6])[0])
    except:
        print('Ошибка занесения в список смешанный')

    try:
        if 'город' in data[7]:
            city.append(regex_num.findall(data[7])[0])
    except:
        print('Ошибка занесения в список город')

    try:
        if 'трасса' in data[8]:
            road.append(regex_num.findall(data[8])[0])
    except:
        print('Ошибка занесения в список трасса')
        
    try: 
        if (float(data[9])>= 0.5 and float(data[9]) <=30):
            do100.append(data[9])
        if (float(data[3])>= 0.5 and float(data[3]) <=30):
            do100.append(data[3])
        if (float(data[7])>= 0.5 and float(data[7]) <=30):
            do100.append(data[7])
    except:
        print('Ошибка занесения в список секунд до 100')
    
    try: 
        if (float(data[10])>= 2015 and float(data[10]) <=2021):
            year.append(data[10])
        if (float(data[4])>= 2015 and float(data[4]) <=2021):
            year.append(data[4])
        if (float(data[8])>= 2015 and float(data[8]) <=2021):
            year.append(data[8])
    except:
        print('Ошибка занесения в список года')
```

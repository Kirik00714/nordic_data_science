# Задача

Вам дан файл `beauty.csv`, в котором собраны несколько факторов про различных людей (мужчин и женщин): зарплата, опыт, цвет кожи, здоровье и т.д.

В последней колонке - целевое значение, которое требуется предсказать. Это то, как выглядит человек (красота) по шкале от 1 до 5. Принимаем, что это задача регрессии, т.е. на выходе могут получаться предикты вроде 3.6

Сначала нужно считать файл и убедиться, что все корректно сохранилось в таблицу.

```python
# your code himport numpy as np
import pandas as pd
import numpy as np
from scipy import optimize
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import (LinearRegression, Ridge, Lasso,
                                    LogisticRegression)
```

```python
df_1 = pd.read_csv('beauty.csv', sep=';')
df_1.head()
```

#### Разделить выборку на train и test с помощью sklearn

А потом убедиться, что распределения таргетов в каждой выборке схожи.

```python
from sklearn.model_selection import  train_test_split
```

```python
X = df_1.values[:, :-1]
y = df_1.values[:, -1]
X.shape,y.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y,
                                                 test_size=0.33,
                                                 random_state=42,
                                                 shuffle = True)
```

```python
y_train , y_test
```

```python
sns.kdeplot(y_train, label='train');
sns.kdeplot(y_test, label='test');
plt.legend();
plt.show();
```

#### Вывести с помощью barh значимость факторов

Что влияет сильнее всего? Слабее всего?

#### Посчитать ошибку (какую сочтете нужным)

```python
from sklearn.metrics import mean_absolute_error, SCORERS, mean_squared_error
```

```python
predicts_test = lr.predict(X_test)
predicts_train = lr.predict(X_train)
```

```python
test_mae = mean_absolute_error(y_test, predicts_test)
test_mae
```

```python
train_mae = mean_absolute_error(y_train, predicts_train)
train_mae
```

```python
print(abs(1-(test_mae/train_mae))*100,'%')
```

```python
mean_squared_error(y_test, predicts_test), mean_squared_error(y_train, predicts_train)
```

```python
test_rmse = np.sqrt(mean_squared_error(y_test, predicts_test))
test_rmse
```

```python
train_rmse = np.sqrt(mean_squared_error(y_train, predicts_train))
train_rmse
```

```python
print(abs((1-(test_rmse/train_rmse))*100),'%')
```

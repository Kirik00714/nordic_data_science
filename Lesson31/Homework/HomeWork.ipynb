```python
class MyGradientBoostingRegressor(object):
    
    def __init__(self, n_estimators=100, max_depth=4, 
                 min_samples_split=2, learning_rate=0.01):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.learning_rate = learning_rate
        self.DecisionTreeRegressorArr = []
        self.train_score_ = []
        
    def my_fit(self, X, y):
        params = {'max_depth': self.max_depth, 
                  'min_samples_split': self.min_samples_split}
        e = y
        pred = np.zeros(len(y))
        for i in range(self.n_estimators):
            dt = DecisionTreeRegressor(**params)
            dt.fit(X, e)
            pred_current = np.array(dt.predict(X))
            pred += pred_current
            self.train_score_.append(((2*(y - pred) + 2*(np.sin((y - pred)) + 1/2)*np.cos((y - pred))) * self.learning_rate))
            e = 2 * self.learning_rate * (y - pred)
            self.DecisionTreeRegressorArr.append(dt)
            
    def my_predict(self, X):
        pred = np.zeros(len(X))
        for dt in self.DecisionTreeRegressorArr:
            pred += np.array(dt.predict(X))
        return pred
```

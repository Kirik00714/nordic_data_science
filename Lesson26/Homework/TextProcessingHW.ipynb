# Обработка текстов

Вам дан файл `ru.conversations.txt`

В нем содержатся диалоги (разделены пустой строкой).

Вам необходимо:

### Разбить текст на диалоги

```python
import nltk
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
from tqdm.auto import tqdm
from nltk.tokenize import RegexpTokenizer
from nltk.tokenize import word_tokenize
nltk.download('punkt')
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
with open("ru.conversations.txt", "r",encoding='utf-8') as file:
    text = file.read()
text = text.replace('"','').replace('-','').replace(',','').\
    replace('.','').replace('!','').replace('?','').lower().split('\n\n')
```

```python
text_1 = [dialog.replace('\n', '') for dialog in text]
```

```python
text_1
```

### Разбить диалоги на токены

```python

tokens = [dialog.split() for dialog in text_1]
    
```

### Предобработать токены
Удалить стоп-слова и спецсимволы, лемматизировать или стеммировать слова и т.д.

```python
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('russian')+['.', ',','-','?','!'])
```

```python
clean_tokens = tokens[:]
for token in tqdm(tokens[:]):
    for word in token:
        if word in stop_words:
            token.remove(word)
```

```python
clean_tokens
```

```python
list_dialog = []
for dialog in clean_tokens:
    list_dialog.append(' '.join(dialog))
```

##### Задание 1. 
В зале суда есть 5 присяжных, каждый из них по отдельности с вероятностью 70% может правильно определить, виновен подсудимый или нет. С какой вероятностью они все вместе вынесут правильный вердикт, если решение принимается большинством голосов?

* 70.00%
* 83.20%
* 83.70%
* 87.50%

Теперь перейдем непосредственно к машинному обучению.

Формула

`N` - количество присяжных

`p` - вероятность правильного решения присяжного

`u` - вероятность правильного решения всего жюри

`m` - минимальное большинство членов жюри, `m = floor(N / 2) + 1`

`C(i,N)` - количество сочетаний из `N` по `i`

`u = sum([C(i,N) * p**i * (1 - p)**(N - i) for i in range(m, N)])`

```python
import math
def C(i,N):
    return math.factorial(N)/(math.factorial(N - i)*math.factorial(i))

N = 5
ver = [0.7, 0.832, 0.837, 0.875]
m = math.floor(N / 2) + 1

for p in ver:
    u = sum([C(i,N) * p**i * (1 - p)**(N - i) for i in range(m, N)])
    print(u)
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = 10, 6
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.ensemble import BaggingClassifier, BaggingRegressor
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
df = pd.read_csv('credit_scoring_sample.csv', sep=';')
df = df.fillna(0)
```

```python
df.head()
```

```python
fig = sns.kdeplot(df[df['SeriousDlqin2yrs'] == 0]['age'], label = 'pay out')
fig = sns.kdeplot(df[df['SeriousDlqin2yrs'] == 1]['age'], label = 'not paying') 
fig.legend(['pay out', 'not paying'])
fig.set(xlabel='Возраст', ylabel='Плотность')    
plt.show()
```

```python
def get_bootstrap_samples(data, n_samples):
    indices = np.random.randint(0, len(data), 
                                (n_samples, len(data)))
    samples = data[indices]
    return samples

def stat_intervals(stat, alpha):
    boundaries = np.percentile(stat, [100 * alpha / 2., 
                                      100 * (1 - alpha / 2.)])
    return boundaries
```

```python
pay_out = df[df['SeriousDlqin2yrs'] == 0]['age'].values
not_paying = df[df['SeriousDlqin2yrs'] == 1]['age'].values
```

```python
# ставим seed для воспроизводимости результатов
np.random.seed(0)

# генерируем выборки с помощью бутстрэра и сразу считаем по каждой из них среднее
pay_out_mean_scores = [np.mean(sample) 
                       for sample in get_bootstrap_samples(pay_out, 1000)]
not_paying_mean_scores = [np.mean(sample) 
                       for sample in get_bootstrap_samples(not_paying, 1000)]

#  выводим интервальную оценку среднего
print("Age from pay out:  mean interval",  stat_intervals(pay_out_mean_scores, 0.05))
print("Age from not paying:  mean interval",  stat_intervals(not_paying_mean_scores, 0.05))
```

В итоге мы получили, что с 95% вероятностью средний возраст тех, кто выплатит будет лежать в промежутке между 52.5 и 52.9, в то время как возраст тех, кто не выплатил в среднем от 45.6 до 46.2 раз.

##### Задание 3.

Подберите оптимальные параметры доли сэмпла и количества деревьев для минимизации MSE.


```python
df_train,df_test = train_test_split(df,test_size=0.25,random_state=17)
```

```python
df_train.head()
```

```python
x_train,y_train = df_train[['age', 'NumberOfTime30-59DaysPastDueNotWorse', 'DebtRatio', 'NumberOfTimes90DaysLate', 'NumberOfTime60-89DaysPastDueNotWorse', 'MonthlyIncome', 'NumberOfDependents']],df_train[['SeriousDlqin2yrs']]

x_test,y_test = df_test[['age', 'NumberOfTime30-59DaysPastDueNotWorse', 'DebtRatio', 'NumberOfTimes90DaysLate', 'NumberOfTime60-89DaysPastDueNotWorse', 'MonthlyIncome', 'NumberOfDependents']],df_test[['SeriousDlqin2yrs']]
```

```python


rf = RandomForestRegressor(n_estimators=10).fit(x_train, y_train)
rf_predict = rf.predict(x_test)

print("Случайный лес, MSE = %.2f" % mean_squared_error(y_test, rf_predict))
```

```python
from tqdm.auto import tqdm


test_acc = []

trees_grid = [5, 10, 15, 20, 30, 50, 75, 100, 125, 150]

min_leaf = [j for j in range(1,10,1)]


for ntrees in tqdm(trees_grid):
    for sampl in min_leaf:
        rfc = RandomForestRegressor(n_estimators=ntrees,
                                   min_samples_leaf=sampl).fit(x_train, y_train)
        rf_predict = rf.predict(x_test)
        test_acc.append(mean_squared_error(y_test, rf_predict))

print("Best MSE is {:.4f}% with {} trees {} samples".format(max(test_acc), 
                                                        trees_grid[np.argmax(test_acc)],
                                                        min_leaf[np.argmax(test_acc)] ))
```

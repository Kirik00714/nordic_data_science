```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import io
import matplotlib.pyplot as plt
```

```python
with io.open('tale.txt',encoding = 'utf-8') as f:
    txt = f.read()
print(txt)
```

```python
txt = txt.lower().replace('!',".").replace('?','.').replace('\n','.').replace(' — ','.').replace(',','').replace('…','.').replace('..','.').replace(';','.').replace(':','.')
```

```python
txt_ = txt.split('.')
df = pd.DataFrame(txt_,columns=["Sentences"])
df
```

```python
glas = ["а", "о", "э", "е", "и", "ы", "у", "ё", "ю", "я"]
count_glas = 0
count_buk = 0
for j, i in df.iterrows():
    for z in i['Sentences']:
        count_buk +=1
        for buk in glas:
            if z == buk:
                count_glas += 1
print(f'{count_glas} гласных')
print(f'{count_buk - count_glas} согласных')
count_sogl = count_buk - count_glas
```

```python
data_ = {'Гласных': [count_glas], 'Согласных': [count_sogl]}
data_
df_ = pd.DataFrame(data_)
df_.plot(kind='bar',rot=0, title = 'Кол-во букв')
plt.ylabel('Кол-во гласных/согласных');
```
